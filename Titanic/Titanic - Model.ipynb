# Titanic - Model
## Data loading and quick snapshot

```python
# CSV reading
import pandas as pd
df_train = pd.read_csv("train.csv", delimiter=',')
df_test = pd.read_csv("test.csv", delimiter=',')
print("Training dataset dimension : {}".format(df_train.shape))
print("Test dataset dimension : {}".format(df_test.shape))
```

```python
# Data combination
targets = df_train['Survived']
df_train.drop(['Survived'], 1, inplace=True)
N_train = df_train.shape[0]
df_combined = df_train.append(df_test)
df_combined.reset_index(inplace=True)

#Data description
print("Data description :")
print(df_combined.dtypes)
print(df_combined.describe(include="all"))
print("\n")
# Missing data
print("Missing data :")
print(round(df_combined.isnull().sum()*100/len(df_combined),2).sort_values(ascending=False))
print("Combined dataset dimension : {}".format(df_combined.shape))
```

## Feature engineering

### Age

```python
# We fill NA values with the mean
df_combined['Age'].fillna((df_combined['Age'].mean()), inplace=True)
```

### Title

```python
# We are going to extract the title of each passenger
df_combined['Title'] = df_combined.Name.str.extract(' ([A-Za-z]+)\.', expand=False) #All letters before the .

# We replace titles with a more common title or as Rare
df_combined['Title'] = df_combined['Title'].replace(['Lady', 'Countess','Capt', 'Col','Don', 'Dr','Major', 'Rev', 'Sir', 'Jonkheer', 'Dona'], 'Rare')
df_combined['Title'] = df_combined['Title'].replace('Mlle', 'Miss')
df_combined['Title'] = df_combined['Title'].replace('Ms', 'Miss')
df_combined['Title'] = df_combined['Title'].replace('Mme', 'Mrs')

# We encode the feature and we drop the initial column
title_dummies = pd.get_dummies(df_combined['Title'], prefix='Title')
df_combined = pd.concat([df_combined, title_dummies], axis=1)
df_combined.drop('Title', axis=1, inplace=True)
```

### Fare

```python
# We fill NA values with the mean
df_combined['Fare'].fillna((df_combined['Fare'].mean()), inplace=True)
```

### Embarked

```python
# We fill NA values with most frequent value
embarked_default = df_combined['Embarked'].value_counts().idxmax()
df_combined['Embarked'].fillna(embarked_default, inplace=True)

# We encode the feature and we drop the initial column
embarked_dummies = pd.get_dummies(df_combined['Embarked'], prefix='Embarked')
df_combined = pd.concat([df_combined, embarked_dummies], axis=1)
df_combined.drop('Embarked', axis=1, inplace=True)
```

### Cabin/Deck

```python
# We create a new feature : deck as the first letter of the cabin
df_combined['Deck']=df_combined['Cabin'].astype(str).str[0]
df_combined.drop(['Cabin'], inplace=True, axis=1)

# We encode the feature and we drop the initial column
deck_dummies = pd.get_dummies(df_combined['Deck'], prefix='Deck')
df_combined = pd.concat([df_combined, deck_dummies], axis=1)
df_combined.drop('Deck', axis=1, inplace=True)
```

### Sex

```python
# We encode the feature and we drop the initial column
sex_dummies = pd.get_dummies(df_combined['Sex'], prefix='Sex')
df_combined = pd.concat([df_combined, sex_dummies], axis=1)
df_combined.drop('Sex', axis=1, inplace=True)
```

### PClass

```python
# We encode the feature and we drop the initial column
pclass_dummies = pd.get_dummies(df_combined['Pclass'], prefix='Pclass')
df_combined = pd.concat([df_combined, pclass_dummies], axis=1)
df_combined.drop('Pclass', axis=1, inplace=True)
```

### Relatives

```python
# We create a new feature : total number of relatives
df_combined['Relative']=df_combined['SibSp']+df_combined['Parch']
df_combined.drop(['SibSp'], inplace=True, axis=1)
df_combined.drop(['Parch'], inplace=True, axis=1)
```

### Index/PassengerID/Ticket/Name

```python
# Index/PassengerID/Ticket/Name
df_combined.drop(['index', 'PassengerId', 'Ticket','Name'], inplace=True, axis=1)
```

### Review of the transformed dataset

```python
print("Data :")
print(df_combined.head())
print(df_combined.dtypes)
print("Missing data :")
print(round(df_combined.isnull().sum()*100/len(df_combined),2).sort_values(ascending=False))
```

## Model

```python
from sklearn.model_selection import train_test_split

# Building of the train and test datasets
train = df_combined.iloc[:N_train]
test = df_combined.iloc[N_train:]
X_train, X_test, y_train, y_test = train_test_split(train, targets, test_size=0.3, random_state=5)
print(X_train)
```

### Learning

```python
# We test several classification methods
Models = []

# SVM Linear
from sklearn import svm
svm_lin = svm.SVC(kernel='linear', C=1.)
Models.append(svm_lin)

# SVM RBF
svm_rbf = svm.SVC(kernel='rbf')
Models.append(svm_rbf)

# Gaussian NB
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
Models.append(gnb)

# Multinomial NB
from sklearn.naive_bayes import MultinomialNB
mnb = MultinomialNB()
Models.append(mnb)

# Gaussian Process
from sklearn.gaussian_process import GaussianProcessClassifier
gp = GaussianProcessClassifier()
Models.append(gp)

# NN
from sklearn.neural_network import MLPClassifier
nn = MLPClassifier(hidden_layer_sizes=(25),solver='sgd', max_iter=2000, learning_rate_init=0.01)
Models.append(nn)

# Decision Tree
from sklearn import tree
dt = tree.DecisionTreeClassifier(criterion='entropy',max_depth=10)
Models.append(dt)

# Boosted Forest
from sklearn.ensemble import AdaBoostClassifier
boosted_forest = AdaBoostClassifier(tree.DecisionTreeClassifier(criterion='entropy',max_depth=10), n_estimators=100)
Models.append(boosted_forest)

# Random Forest
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, criterion='entropy')
Models.append(rf)
```

```python
import random
import numpy as np
from sklearn.model_selection import StratifiedKFold

# Cross validation for each model
def learn(models, N):
    Kfold = StratifiedKFold(n_splits=N, shuffle=True)
    for model in models:
        print(model)
        score = []
        for train_index, test_index in Kfold.split(train, targets):
            Xtrain, Xtest = train.loc[train_index], train.loc[test_index]
            ytrain, ytest = targets.loc[train_index], targets.loc[test_index]
            model.fit(Xtrain,ytrain);
            score += [model.score(Xtest,ytest)]
            print('*', end='')
        print(" done!")
        print("Average generalization score:", np.mean(score))
        print("Standard deviation:", np.std(score))
        print()
        
learn(Models,4)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# Hyperparameters of SVM tuning using GridSearchCV
parameters = {'kernel':('linear', 'rbf'), 'C':np.logspace(-4,0,5)}
svc = svm.SVC()
clf = GridSearchCV(svc, parameters)
clf.fit(X_train, y_train)

print("Best parameters set found on development set:")
print()
print(clf.best_params_)
print()
print("Grid scores on development set:")
print()
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r"
          % (mean, std * 2, params))
print()

print("Detailed classification report:")
print()
print("The model is trained on the full development set.")
print("The scores are computed on the full evaluation set.")
print()
y_true, y_pred = y_test, clf.predict(X_test)
print(classification_report(y_true, y_pred))
print()
```

```python
from sklearn import tree
from os import system
from IPython.display import Image

# Display of the decision tree
dt = tree.DecisionTreeClassifier(criterion='entropy',max_depth=10)
dt.fit(X_train,y_train)
print(dt.score(X_test,y_test))

def disp_tree(filename, treename):
    dotfile = open(filename+'.dot', 'w')
    tree.export_graphviz(treename, 
                         out_file = dotfile,
                         filled=True,
                         rounded=True,  
                         special_characters=True)
    dotfile.close()
    system("dot -Tpng "+filename+".dot -o "+filename+".png")
    return Image(filename+'.png')

disp_tree('dt',dt)
```

```python
# Hyperparameters of Boosted Forest tuning using GridSearchCV
parameters = {'max_depth':np.arange(1,16,1), 'min_samples_leaf':np.arange(1,30,1)}
clf = AdaBoostClassifier(tree.DecisionTreeClassifier(criterion='entropy'),n_estimators=100)
clf = GridSearchCV(dt, parameters)
clf.fit(X_train, y_train)

print("Best parameters set found on development set:")
print()
print(clf.best_params_)
print()
print("Grid scores on development set:")
print()
means = clf.cv_results_['mean_test_score']
stds = clf.cv_results_['std_test_score']
for mean, std, params in zip(means, stds, clf.cv_results_['params']):
    print("%0.3f (+/-%0.03f) for %r"
          % (mean, std * 2, params))
print()

print("Detailed classification report:")
print()
print("The model is trained on the full development set.")
print("The scores are computed on the full evaluation set.")
print()
y_true, y_pred = y_test, clf.predict(X_test)
print(classification_report(y_true, y_pred))
print()
```
