# Digit Recognizer 
## Data Exploration : Train Dataset
### Training Data Loading

```python
# CSV reading
import pandas as pd
import numpy as np
df_train = pd.read_csv("train.csv", delimiter=',')

# Training labels
labels_train=pd.DataFrame()
labels_train['label']=df_train.label

# Training images
images_train=pd.DataFrame()
images_train=df_train.copy()
images_train.drop("label",axis=1, inplace=True)
```

### Class Balance

```python
# Count of each label occurences
df_train.groupby(by='label').count()['pixel0']
pd.pivot_table(df_train,index=["label"],aggfunc='count')
```

### Conversion to Numpy

```python
# df -> np
labels_train=labels_train.to_numpy()
labels_train=labels_train[:,0]
images_train=images_train.to_numpy()
```

### Visualization

```python
import matplotlib.pyplot as plt
import random

def display_image(images,labels, rows, columns, preds=None):
    
    fig=plt.figure(figsize=(15, 15))
    
    for k in range(1, columns*rows +1):
            
        i = random.randint(0, len(images))

        image=images[i]
        label=labels[i]
        if preds is not None:
            pred=preds[i]

        image = image.reshape((1,28,28))
        image = image[0,:,:]
        
        fig.add_subplot(rows, columns, k)
        plt.imshow(image, cmap="gray")
        if preds is not None:
            pred=preds[i]
            plt.title("Label : " + str(label) + " - Prediction : " + str(pred))
        else:
            plt.title("Label : " + str(label))

display_image(images_train,labels_train,4,4)
```

## Model

```python
from sklearn.model_selection import train_test_split

# Normalization of the images
images_train = images_train/255.

X_train, X_test, y_train, y_test = train_test_split(images_train, labels_train, test_size=0.1, random_state=0)

# One hot encoding
from keras.utils import to_categorical
y_train_ohe = to_categorical(y_train)
y_test_ohe = to_categorical(y_test)
```

### Deep Learning

```python
import keras.metrics as metrics

def top3_acc(ytrue, ypred):
    return metrics.top_k_categorical_accuracy(ytrue, ypred, k=3)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Dropout, Flatten, MaxPooling2D
import keras.losses as losses
import keras.optimizers as optimizers

# Model architecture
model = Sequential()
model.add(Conv2D(28, kernel_size=(3,3), input_shape=(28, 28, 1)))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Flatten())
model.add(Dense(128, activation=tf.nn.relu))
model.add(Dropout(0.2))
model.add(Dense(10,activation=tf.nn.softmax))

# Learning parameters
opt = optimizers.Adam(learning_rate=0.001)
loss = losses.categorical_crossentropy

# Model compilation
model.compile(optimizer=opt, loss=loss, metrics = [metrics.categorical_accuracy, top3_acc])
```

```python
# Display the model
from keras.utils import plot_model
plot_model(model, to_file='model.png')
```

```python
%%time
history = model.fit(X_train.reshape(-1, 28, 28, 1), y_train_ohe, validation_data=(X_test.reshape(-1, 28, 28, 1),y_test_ohe), epochs=20, batch_size=1024)
```

```python
plt.figure(figsize=(20,10))
plt.style.use('ggplot')
for i, (name, values) in enumerate(sorted(history.history.items())):
    plt.subplot(1, len(history.history.items()), i+1)
    plt.plot(values)
    plt.title(name)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix

# Prediction on the test images
y_pred_ohe = model.predict(X_test.reshape(-1, 28, 28, 1))
y_pred = y_pred_ohe.argmax(axis=1) # ohe -> label

# Confusion matrix
print("Confusion matrix :")
print(confusion_matrix(y_test, y_pred))

# Images missclassified
X_mclf = X_test[y_pred != y_test]
y_true_mclf = y_test[y_pred != y_test]
y_pred_mclf = y_pred[y_pred != y_test]
display_image(X_mclf,y_true_mclf, 4, 4, y_pred_mclf)

# Images well classified
X_clf = X_test[y_pred == y_test]
y_true_clf = y_test[y_pred == y_test]
y_pred_clf = y_pred[y_pred == y_test]
display_image(X_clf,y_true_clf, 4, 4, y_pred_clf)
```

## Submission

```python
# CSV reading
df_test = pd.read_csv("test.csv", delimiter=',')

# Data processing
images_test = df_test.to_numpy()
images_test = images_test/255.
```

```python
# Prediction
labels_pred_ohe = model.predict(images_test.reshape(-1, 28, 28, 1))
labels_pred = labels_pred_ohe.argmax(axis=1)
```

```python
# Formatting of the output
dflabels_pred = pd.DataFrame({'Label': labels_pred})
dfid = pd.DataFrame({'ImageId': np.arange(1,len(labels_pred)+1,1).astype(int)})
dfsubmission = pd.concat([dfid,dflabels_pred],axis=1)

# CSV creation
dfsubmission.to_csv('ConvNN.csv', index = False)
```
